Since formula for the Straight line is given by:-

$$ Y = m*X +B $$

where
 **m = Slop** , **B = Intercept**

 Formula for calculating **Slop** and **Intercept** is given by:-

$$ m = \frac{\bar{X}.\bar{Y} - \bar{X.Y}}{\bar{X}^2 - \bar{X^2}}$$

and

$$ B = \bar{Y} - m.\bar{X}$$

where
$\bar{X}$ = Mean of X,
$\bar{Y}$ = Mean of Y

```python
from statistics import mean
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')

xs = np.array([1,2,3,4,5,6], dtype= np.float64)
ys = np.array([5,4,6,5,6,7], dtype = np.float64)

def best_fit_slop_and_intercept(xs,ys):
    '''
    Function for calculating the slop and intercept of the data
    '''

    m = ( ( ( mean(xs) * mean(ys) ) - mean(xs*ys) )  / ( mean(xs)**2 - mean(xs**2) ))
    b = mean(ys) - m * mean(xs)
    return m , b


m ,b = best_fit_slop_and_intercept(xs,ys)
print('Slop = ',m ,'\nIntercept = ', b)

regression_line = [(m*x) + b for x in xs]

predict_x = 8
predict_y = (m*predict_x) + b



plt.scatter(xs , ys)
plt.scatter(predict_x, predict_y, color= 'r', label = "predicted values")
plt.plot(xs , regression_line)
plt.show()
```

## R Squared Error ($R^2$ Error)
In statistics, the coefficient of determination, denoted $R^2$ or $r^2$ and pronounced "R squared", is the proportion of the variance in the dependent variable that is predictable from the independent variable(s).

A data set has n values marked y1,...,yn (collectively known as yi or as a vector y = $[y1,...,yn]^T$), each associated with a fitted (or modeled, or predicted) value f1,...,fn (known as fi, or sometimes ŷi, as a vector f).

**Defining  the residuals as $e_i$ = $y_i$ − $f_i$ (forming a vector $\vec{e}$).**

If $\bar{y} $ is the mean of the observed data:

$ \bar{Y} = \frac{1}{n}\sum_{i=0}^n{Y}$

**Then the variability of the data set can be measured with two sums of squares formulas:**

- The total sum of square (proportional to the variance of the data)

$ SS_{tot} =\sum_i (Y_i - \bar{Y})^2 $

- The sum of square of residuals, also called the residual sum of squares

$ SS_{res} = \sum_i (Y_i - F_i)^2$

The most general definition of the co-efficient of determination 

$ R^2 = 1- \frac{SS_{res}}{SS_{tot}} $

In the best case, the modeled value exactly match the observed vlaues which results in $ SS_[res]$ = 0 , and $R^2$ =1
A baseline model which always predicts $\bar{Y}$, will have $R^2$ =0 models that have worse prediction than this baseline will have a negative $R^2$.

```python
def squared_error(ys_orig , ys_line):
    return sum((ys_orig - ys_line)**2)

def coefficient_of_determination (ys_orig , ys_line):
    ys_mean_orig = [mean(ys_orig) for y in ys_orig]
    total_squared_error = squared_error(ys_orig , ys_mean_orig)
    residual_squared_error = squared_error(ys_orig , ys_line)

    return (1 - (residual_squared_error / total_squared_error))

r_squared = coefficient_of_determination(ys , regression_line)
print(r_squared)
```
